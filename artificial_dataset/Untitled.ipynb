```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
import pandas as pd
from pandas import DataFrame
import seaborn as sns
```

```python
df = pd.read_fwf('output(5000000).txt')
df.to_csv('op_conv.csv')
df.drop(["Max", "Test"], axis=1)
```

```python
sns.lmplot( x="Depth", y="Accuracy", data=df, fit_reg=False)
```

```python
sns.lmplot( x="size", y="Accuracy", data=df, fit_reg=False)
```

```python
cmap = sns.cubehelix_palette(as_cmap=True)

f, ax = plt.subplots()
points = ax.scatter(df["Depth"], df["Training Time"], c=df["Accuracy"], s=50, cmap=cmap)
f.colorbar(points)

#sns.lmplot( x="Depth", y="Training Time",hue="Accuracy", data=df, fit_reg=False)
```

```python
cmap = sns.cubehelix_palette(as_cmap=True)

f, ax = plt.subplots()
points = ax.scatter(df["size"], df["Training Time"], c=df["Accuracy"], s=50, cmap=cmap)
f.colorbar(points)
```
