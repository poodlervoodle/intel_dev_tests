```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import ListedColormap 
import statistics
cmap = ListedColormap(sns.color_palette("mako_r", 256).as_hex())
```

```python
f = pd.read_fwf('samples_features_nonintel.txt')
f.to_csv('samples_features_nonintel.csv')
f = f.drop(["Unnamed: 0", "Test", "Max"], axis = 1)
```

```python
x = f['Features']
y = f['Samples']
z = f['Accuracy']
t = f['Training Time']
print("Accuracy :", statistics.mean(z), statistics.stdev(z))
print("Training Time :", statistics.mean(t), statistics.stdev(y))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')

# get colormap from seaborn
#cmap = ListedColormap(sns.color_palette("husl", 256).as_hex())

# plot
sc = ax.scatter(x, y, z, s=40, c=f['Training Time'], marker='o', cmap=cmap, alpha=1)
ax.set_xlabel('Features')
ax.set_ylabel('Samples')
ax.set_zlabel('Accuracy')

# legend
#plt.legend(*sc.legend_elements(), bbox_to_anchor=(1.05, 1), loc=2)
plt.colorbar(sc)
```

```python
points = plt.scatter(x, z,
                     c=t, s=20, cmap=cmap)
plt.colorbar(points)
plt.xlabel('Features') 
plt.ylabel('Accuracy')
```

```python
points = plt.scatter(y, z,
                     c=t, s=20, cmap=cmap)
plt.colorbar(points)
plt.xlabel('Samples') 
plt.ylabel('Accuracy')
```

```python
points = plt.scatter(x, t,
                     c=z, s=20, cmap=cmap)
plt.colorbar(points)
plt.xlabel('Features') 
plt.ylabel('Training Time')
```

```python
points = plt.scatter(y, t,
                     c=z, s=20, cmap=cmap)
plt.colorbar(points)
plt.xlabel('Samples') 
plt.ylabel('Training Time')
```
