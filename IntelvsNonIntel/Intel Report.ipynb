```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import ListedColormap 
import statistics
```

```python
cmap = ListedColormap(sns.color_palette("husl", 256).as_hex())
```

```python
f = pd.read_fwf('depth_intel.txt')
f.to_csv('depth_intel.csv')
```

```python
x = f['Depth']
y = f['Accuracy']
t = f['Training Time']
print("Accuracy :", statistics.mean(y))
print("Training Time :", statistics.mean(t))
```

```python
points = plt.scatter(x, y,
                     c=t, s=20, cmap=cmap)
plt.colorbar(points)
plt.xlabel('Depth') 
plt.ylabel('Accuracy')
```

```python
points = plt.scatter(x, t,
                     c=x, s=20, cmap=cmap)
plt.colorbar(points)
plt.xlabel('Depth') 
plt.ylabel('Training Time')
```
